# **INTRODUÇÃO** 

Com o passar dos anos, o avanço da medicina e acesso aos direitos humanos (vida, alimentação, assistencia médica, serviços sociais), a população mundial apresenta maior expectativa de vida. O estudo exploratório a seguir tem como objetivo mapear as nações com maior e menor expectativa de vida e investigar possíveis razões.


---

```python
import numpy as np # Importação de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv ('https://raw.githubusercontent.com/fabiomiguell/AtividadesMackenzie/main/life_expectancy_years.csv')#expectativa de vida mundial
df.head()
```

```python
df.shape
```

#**Quais são os 10 países com maior expectativa de vida em 2018**?

---




```python
maiores_2018 = df[['country', '2018']].nlargest(10, '2018')
maiores_2018.head(10)
```


#**Quais são os 10 países com menor expectativa de vida em 2018**?



---



```python
menores_2018 = df[['country', '2018']].nsmallest(10, '2018')
menores_2018.head(10)
```

#**Considerando as 10 nações com menor expectativa de vida, foi selecionado o país República Centro-Africana para estudo exploratório**
---

#**1- Acesso a água**

```python
df_2 = pd.read_csv ('https://raw.githubusercontent.com/fabiomiguell/AtividadesMackenzie/main/at_least_basic_water_source_overall_access_percent.csv')#aquisição dos dados
df_2.head()
```

#**1.1- Quais são os 10 países com menos acesso a água potável?**

```python
acesso_agua_menores_2018 = df_2[['country', '2018']].nsmallest(10, '2018')
acesso_agua_menores_2018.head(10)
```

```python
fig = plt.figure(figsize=(20, 10))
sns.barplot(
    data=acesso_agua_menores_2018,
    x=acesso_agua_menores_2018['country'],
    y=acesso_agua_menores_2018['2018'],
    palette='magma'
    )    
 
```
